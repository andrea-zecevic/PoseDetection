##Uvod <br>
Detekcija poza je napredna tehnika u području računalnog vida i strojnog učenja koja se fokusira na identifikaciju i praćenje ljudskih poza i pokreta u stvarnom vremenu. Ova tehnologija omogućuje računalima da interpretiraju ljudsku figuru, identificirajući ključne točke na tijelu, poput zglobova i udova, i prateći njihov položaj i kretanje.<br><br>

U sportu, detekcija poza otvara nove mogućnosti za trenere i sportaše, omogućavajući detaljnu analizu pokreta i tehniku, što može biti ključno za poboljšanje performansi i smanjenje rizika od ozljeda. <br><br>

U kontekstu tenisa, detekcija poza omogućava igračima i trenerima da analiziraju i optimiziraju udarce npr. servis. Ovo im pomaže u poboljšanju tehnike udarca i u prevenciji ozljeda, jer precizna analiza može otkriti potencijalne probleme u biomehanici igrača.

##Pregled i implementacija <br>
Detekcija poza uključuje prepoznavanje i analizu ljudskih figura u slikama ili videozapisima. Osnovni princip ovog procesa je identificiranje ključnih točaka na ljudskom tijelu, poput zglobova ruku, nogu, i glave. Ove točke se zatim koriste za stvaranje "skeleta" koji predstavlja pozu subjekta. <br> <br>

Najpoznatija i najčešće korištena metoda za detekciju poza temelji se na upotrebi **dubokih neuronskih mreža (Deep Neural Networks - DNNs)**.

Osim DNN-ova, postoje i druge metode detekcije poza koje se temelje na različitim tehničkim pristupima, kao što su:

- **Tradicionalne metode računalnog vida:** Ove metode koriste tradicionalne tehnike obrade slike i računalnog vida, poput detekcije rubova, HOG (histograma usmjerenih gradijenata) značajki, i optičkog toka za izdvajanje informacija iz slike i prepoznavanje poza.

- **Detekcija poza pomoću senzora:** U nekim aplikacijama, poput praćenja pokreta u igrama, koriste se senzori poput Kinecta ili akcelerometara kako bi se dobile informacije o pozi korisnika.

- **Optičko praćenje pokreta:** Ova metoda koristi kamere za praćenje pokreta osobe i analizu promjena u pozama tijekom vremena. <br> <br>

Unatoč raznolikosti metoda, duboke neuronske mreže su postale najpopularniji pristup detekciji poza zbog njihove sposobnosti da nauče složene obrasce iz velikih skupova podataka.

###Metode temeljene na dubokom učenju

Metode temeljene na dubokom učenju, posebice konvolucijske neuronske mreže (CNN), predstavljaju revolucionarni pristup u računalnom vidu. Ove tehnike koriste se za analizu slika i videa, gdje se neuronske mreže treniraju na velikim skupovima podataka s precizno označenim položajima ključnih točaka na ljudskom tijelu.

**Proces učenja:**

CNN-ovi prolaze kroz proces učenja gdje mreža automatski uči prepoznati složene uzorke i karakteristike ljudske figure, poput položaja ruku, nogu i glave.

**Prednosti:**

Ove metode nude visoku točnost i sposobnost generalizacije, što ih čini idealnim za kompleksne zadatke u računalnom vidu kao što je detekcija poza. [1]

Kod implementacije detekcije poza u projektima obično razmatramo dva osnovna pristupa: **sami treniramo model od početka ili koristimo već postojeće, unaprijed trenirane modele.** <br><br>

**Treniranje vlastitog modela** <br>
Ovaj pristup uključuje razvoj i treniranje neuronske mreže od nule ili prilagodbu postojećih modela specifičnim potrebama projekta. Glavna prednost ovog pristupa je visoka razina prilagodljivosti; model se može optimizirati za specifične zahtjeve projekta. Međutim, treniranje modela zahtijeva značajne resurse: velike i dobro označene skupove podataka, znatnu računalnu snagu, i stručno znanje u području strojnog učenja. Također, proces treniranja može biti vremenski zahtjevan. <br> <br>


<br><br>
**Korištenje prethodno treniranih, gotovih modela** <br>
Gotovi modeli su već optimizirani i trenirani na opsežnim skupovima podataka, pružajući visoku točnost i pouzdanost bez potrebe za vlastitim treniranjem. Prednosti uključuju brzu implementaciju, manje tehničke složenosti, i smanjenu potrebu za računalnim resursima. Međutim, nedostatak ovog pristupa je manja fleksibilnost u prilagodbi modela specifičnim potrebama.

**Kako funkcionira detekcija poza u MediaPipe-u?**

1. Proces započinje učitavanjem slike ili videozapisa u sustav.

```python
from mediapipe import solutions                         # Pristup unaprijed izgrađenim modelima i alatima koji se koriste za različite zadatke poput detekcije poza
from mediapipe.framework.formats import landmark_pb2    # Rad s podacima o ključnim točkama (landmarks) koje vraćaju MediaPipe modeli
import numpy as np
```

```python
# Funkcija 'draw_landmarks_on_image' prima RGB sliku i rezultate detekcije poze,
# te iscrtava ključne točke detektirane poze na slici.


def draw_landmarks_on_image(rgb_image, detection_result):
  # Izvlači listu detektiranih poznih ključnih točaka iz rezultata detekcije
  pose_landmarks_list = detection_result.pose_landmarks
  # Stvara kopiju ulazne RGB slike na kojoj će se iscrtavati ključne točke
  annotated_image = np.copy(rgb_image)

  # Prolazi kroz sve detektirane poze
  for idx in range(len(pose_landmarks_list)):
    pose_landmarks = pose_landmarks_list[idx]

    # Stvara Protobuf strukturu za ključne točke trenutne poze
    pose_landmarks_proto = landmark_pb2.NormalizedLandmarkList()
    pose_landmarks_proto.landmark.extend([
      # Pretvara svaku ključnu točku u NormalizedLandmark i dodaje ih u listu
      landmark_pb2.NormalizedLandmark(x=landmark.x, y=landmark.y, z=landmark.z) for landmark in pose_landmarks
    ])

    # Koristi MediaPipe funkciju za crtanje ključnih točaka i veza između njih na slici
    solutions.drawing_utils.draw_landmarks(
      annotated_image,
      pose_landmarks_proto,
      solutions.pose.POSE_CONNECTIONS,
      solutions.drawing_styles.get_default_pose_landmarks_style())

  # Vraća sliku s iscrtanim ključnim točkama
  return annotated_image
```

##Detekcija poze na fotografiji

Identifikacija i vizualizacija ključnih točaka poze na fotografiji

```python
import cv2                                    # Biblioteka za obradu slika i videozapisa
import mediapipe as mp                        # Okvir za naprednu analizu slika/videozapisa, uklj. detekciju poza
from google.colab.patches import cv2_imshow   # Omogućava prikaz slika u Google Colab bilježnicama
```

```python
# Kreira se instanca MediaPipe Pose detekcije, postavljena za obradu statičkih slika
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=True)

# Učitavanje slike pomoću OpenCV-ja
image = cv2.imread('/content/tennis_serve_1.jpeg')
# OpenCV učitava sliku u BGR formatu, pa se ona konvertira u RGB format jer MediaPipe radi s RGB formatom
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Detekcija poze, slika se prosljeđuje MediaPipe Pose modelu za detekciju poza
results = pose.process(image_rgb)

# Vizualizacija rezultata
if results.pose_landmarks:
    mp.solutions.drawing_utils.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
    cv2_imshow(image)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```

##Detekcija poze na videozapisu

```python
# Inicijalizacija MediaPipe Pose za obradu videozapisa
# Postavke su prilagođene za kontinuiranu obradu videozapisa

mp_pose = mp.solutions.pose

with mp_pose.Pose(static_image_mode=False, model_complexity=2, min_detection_confidence=0.5) as pose:
    # Učitavanje videozapisa
    cap = cv2.VideoCapture('/content/tennis_video.mp4')

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Konverzija slike u RGB format
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Detekcija poze na svakoj sličici videozapisa
        results = pose.process(frame_rgb)

        # Vizualizacija detektiranih poza na svakoj sličici
        if results.pose_landmarks:
            mp.solutions.drawing_utils.draw_landmarks(
                frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
            cv2_imshow(frame)

        # Prekid petlje s pritiskom na tipku
        if cv2.waitKey(5) & 0xFF == 27:
            break

    # Oslobađanje resursa
    cap.release()

# Rezultat je niz sličica koje prikazuju detektirane poze u različitim trenucima videozapisa
```

U ovoj ćeliji, umjesto da se svaka sličica prikazuje pojedinačno, sličice se zapisuju u novi videzapis.

```python
# Inicijalizacija MediaPipe Pose za obradu videozapisa
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False)

# Učitavanje videozapisa koristeći OpenCV
cap = cv2.VideoCapture('/content/tennis_video.mp4')

# Dobivanje širine i visine videozapisa iz objekta VideoCapture
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Inicijalizacija VideoWriter-a za snimanje izlaznog videozapisa
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter('/content/OUTPUT_tennis_video.mp4', fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break

    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(frame_rgb)

    # Vizualizacija detektiranih poza na slikama
    if results.pose_landmarks:
        mp.solutions.drawing_utils.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Zapisivanje obrađene sličice u izlaznu video datoteku
    out.write(frame)

# Oslobađanje resursa
cap.release()
out.release()
```

##Analiza servisa koristeći ključne točke

```python
def calculate_angle(a, b, c):
    a = np.array(a)  # Prva točka
    b = np.array(b)  # Druga točka (vrh kuta)
    c = np.array(c)  # Treća točka

    # Računa kut između vektora (b, a) i (b, c) koristeći arctan2
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)  # Pretvara radijane u stupnjeve i osigurava da je kut pozitivan

    return angle
```

```python
def calculate_angle_2d(a, b):
    # Izračunava kut između dvije točke (a i b) u 2D prostoru.
    # Koristi se za jednostavnije scenarije gdje su potrebne samo dvije točke.
    a = np.array(a)  # Prva točka
    b = np.array(b)  # Druga točka (vrh kuta)

    radians = np.arctan2(b[1] - a[1], b[0] - a[0])
    angle = (radians * 180.0 / np.pi + 360) % 360

    return angle
```

```python
def extract_landmarks_and_calculate_angles(frame, pose_model):
    results = pose_model.process(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    elbow_angles = []
    knee_angles = []
    body_angles = []
    arm_angles = []

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        # Izračun kuta lakta
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]
        wrist = [landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_WRIST.value].y]

        elbow_angle = calculate_angle(shoulder, elbow, wrist)
        elbow_angles.append(elbow_angle)

        # Izračun kuta koljena
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        knee = [landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_KNEE.value].y]
        ankle = [landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ANKLE.value].y]

        knee_angle = calculate_angle(hip, knee, ankle)
        knee_angles.append(knee_angle)

        # Izračun kuta tijela (kut između ramena i kukova)
        shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]
        hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]

        body_angle = calculate_angle_2d(shoulder, hip)
        body_angles.append(body_angle)

        # Izračun kuta ruke (kut između lakta i ramena)
        elbow = [landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].x, landmarks[mp_pose.PoseLandmark.RIGHT_ELBOW.value].y]

        arm_angle = calculate_angle_2d(shoulder, elbow)
        arm_angles.append(arm_angle)


    return elbow_angles, knee_angles, body_angles, arm_angles
```

```python
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False)

cap = cv2.VideoCapture('/content/tennis_video.mp4')
frame_count = 0

if not cap.isOpened():
    print("Greška pri otvaranju videozapisa.")
else:
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))  # Ukupan broj okvira
    fps = cap.get(cv2.CAP_PROP_FPS)  # FPS videozapisa

timestamps = []
elbow_angles = []
knee_angles = []
body_angles = []
arm_angles = []

with pose as p:
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Pozivanje funkcije za izračun svih kutova
        e_angles, k_angles, b_angles, a_angles = extract_landmarks_and_calculate_angles(frame, p)

        # Dodavanje vremenske oznake i kutova u odgovarajuće liste
        timestamp = frame_count / fps
        timestamps.append(timestamp)
        if e_angles:
            elbow_angles.append(e_angles[0])
        if k_angles:
            knee_angles.append(k_angles[0])
        if b_angles:
            body_angles.append(b_angles[0])
        if a_angles:
            arm_angles.append(a_angles[0])

        frame_count += 1

cap.release()
# Ispis FPS-a i vremenskih oznaka
print("FPS videozapisa:", fps)
print("Vremenske oznake (sekunde):", timestamps)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.plot(timestamps, elbow_angles, label='Kut lakta', color='blue')

# Dodavanje oznaka i naslova
plt.xlabel('Vrijeme (sekunde)')
plt.ylabel('Kut (°)')
plt.title('Promjena kuta desnog lakta tijekom servisa')

# Dodavanje legende
plt.legend()

# Prikaz grafa
plt.show()
```

```python
# Prikaz promjena kutova koljena
plt.figure(figsize=(20, 5))
plt.plot(timestamps, knee_angles, label='Kut koljena', color='blue')

# Dodavanje oznaka i naslova
plt.xlabel('Vrijeme (sekunde)')
plt.ylabel('Kut (°)')
plt.title('Promjene kuta desnog koljena tijekom servisa')

# Dodavanje legende
plt.legend()

# Prikaz grafa
plt.show()
```

```python
# Prikaz promjena kuta ruke
plt.figure(figsize=(20, 5))
plt.plot(timestamps, arm_angles, label='Kut ruke', color='red')

# Dodavanje oznaka i naslova
plt.xlabel('Vrijeme (sekunde)')
plt.ylabel('Kut (°)')
plt.title('Promjene kutova desne ruke tijekom servisa')

# Dodavanje legende
plt.legend()

# Prikaz grafa
plt.show()
```

```python
# Prikaz promjena kuta tijela
plt.figure(figsize=(20, 5))
plt.plot(timestamps, body_angles, label='Kut tijela', color='green')

# Dodavanje oznaka i naslova
plt.xlabel('Vrijeme (sekunde)')
plt.ylabel('Kut (°)')
plt.title('Promjene kuta tijela tijekom servisa')

# Dodavanje legende
plt.legend()

# Prikaz grafa
plt.show()
```
